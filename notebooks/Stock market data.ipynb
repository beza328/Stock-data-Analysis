Explanation of Columns:<br>Date: The trading date. <br>Open: The price at which the stock opened at the start of trading on that day.<br>High: The highest price the stock reached during the trading day.
    <br>Low: The lowest price the stock reached during the trading day.
    <br>Close: The price at which the stock closed at the end of the trading day.
    <br>Adj Close: The closing price adjusted for dividends, stock splits, and other corporate actions.
    <br>Volume: The number of shares traded during the day.

Historical analysis of the seven's company information<br>Although the givent datas are long period datas here I am obligated to sample the last two years data of for clear visualizatioin

```python
# Create a combined DataFrame
combined_data = pd.DataFrame({
    'GOOGL': Goog['Close'],
    'AAPL': Apple['Close'],
    'MSFT': Msft['Close'],
    'NVDA': nvda['Close'],
    'TSLA': tsla['Close'],
    'AMZN': Amzn['Close'],
    'META': Meta['Close'],
 
})

# Display the first few rows to check the combined data
print(combined_data.head())
```

```python
import datetime

# Define the cutoff date for the last 5 years
cutoff_date = datetime.datetime.now() - pd.DateOffset(years=2)

# Filter the combined data for the last 5 years
last_2_years_data = combined_data[combined_data.index >= cutoff_date]

# Display the first few rows of the filtered data
print(last_2_years_data.head())
```

```python
# Plot the stock prices for the last 5 years
plt.figure(figsize=(14, 7))

# Plot each stock's closing price
for ticker in last_2_years_data.columns:
    plt.plot(last_2_years_data.index, last_2_years_data[ticker], label=ticker)

plt.title('Stock Prices Over Time (Last 2 Years)')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

This stock market information shown us 7 company's stock price movement starting from 2022 sept to the july 2024<br>starting from the lowest stock price Nvida has slight change throughout the two years so we can say that it is has stable movement<br>Meta has the most rapid change of stock price increasingly fast until now, it is more likely seems good chance to return back profit in a small period of time but it has no smooth change those fluctuation occure in a short period gap so there need to  be a further study before investing<br>Apple , Amazon, Google and Nvida has Stable change from the previoius progres from this I can say that they have good resistance to factors that make the other company's flactuate in this years.
